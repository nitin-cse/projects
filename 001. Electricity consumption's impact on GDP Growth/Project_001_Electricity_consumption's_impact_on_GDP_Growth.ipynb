## **7. MODEL: Multiple Model Training**

### **7.1. MODEL: Module Importing**

```python
# Importing module for ML Models
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

# Importing module for Regression Prediction Measurement
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

### **7.2. MODEL: Function Definition**

```python
# Model Function Definition

# 1. LinearRegression models without feature selection
def LR(X_train, X_test, y_train, y_test):
      # Training and Predicting model
      logReg = LinearRegression()
      logReg.fit(X_train, y_train)
      y_pred = logReg.predict(X_test)

      # Caculating prediction error metrics
      mae = mean_absolute_error(y_test, y_pred)
      mse = mean_squared_error(y_test, y_pred)
      rem = 100*r2_score(y_test,y_pred)

      print('Mean Absolute Error for Linear Regression: ', mean_absolute_error(y_test, y_pred))
      print('Mean Squared Error for Linear Regression: ', mean_squared_error(y_test, y_pred))
      print('R2 Score for Linear Regression: ', 100*r2_score(y_test, y_pred))
      print()

      return (mae, mse, rem)

# 2.SVM Regression models without feature selection
def SVMR(X_train, X_test, y_train, y_test):
      # Training and Predicting model
      svm_model = SVR()
      svm_model.fit(X_train,y_train)
      y_pred = svm_model.predict(X_test)

      # Caculating prediction error metrics
      mae = mean_absolute_error(y_test, y_pred)
      mse = mean_squared_error(y_test, y_pred)
      rem = 100*r2_score(y_test,y_pred)

      print('Mean Absolute Error for Support Vector Machine: ', mean_absolute_error(y_test, y_pred))
      print('Mean Squared Error for Support Vector Machine: ', mean_squared_error(y_test, y_pred))
      print('R2 Score for Support Vector Machine: ', 100*r2_score(y_test, y_pred))
      print()

      return (mae, mse, rem)

# 3. Random Forest Regression models without feature selection
def RF(X_train, X_test, y_train, y_test):
      # Training and Predicting model
      Random_Forest= RandomForestRegressor()
      Random_Forest= RandomForestRegressor(max_depth=7, random_state=0)
      Random_Forest.fit(X_train,y_train)
      y_pred = Random_Forest.predict(X_test)

      # Caculating prediction error metrics
      mae = mean_absolute_error(y_test, y_pred)
      mse = mean_squared_error(y_test, y_pred)
      rem = 100*r2_score(y_test,y_pred)

      print('Mean Absolute Error for Random Forest: ', mean_absolute_error(y_test, y_pred))
      print('Mean Squared Error for Random Forest: ', mean_squared_error(y_test, y_pred))
      print('R2 Score for Random Forest: ', 100*r2_score(y_test, y_pred))
      print()

      return (mae, mse, rem)
```

### **7.3. MODEL: Training And Evaluation**

```python
# Model Training
(LR_Normal_mae, LR_Normal_mse, LR_Normal_rem) = LR(input_data_train, input_data_test, output_data_train, output_data_test)
(SVMR_Normal_mae, SVMR_Normal_mse, SVMR_Normal_rem) = SVMR(input_data_train, input_data_test, output_data_train, output_data_test)
(RF_Normal_mae, RF_Normal_mse, RF_Normal_rem) = RF(input_data_train, input_data_test, output_data_train, output_data_test)
```

### **8. Optimization**

### **8.1 Optimization technique using Grid Search Cv for Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid
param_grid = {
    'fit_intercept': [True, False],
    'n_jobs': [1, 2, 3, 4],
    'positive': [True, False]
}

# Create a grid search object
grid_search = GridSearchCV(LinearRegression(), param_grid, cv=3)

# Fit the model to the training data
grid_search.fit(input_data_train, output_data_train)

# Get the best model
best_model = grid_search.best_estimator_

# Make predictions on the test data
y_pred = best_model.predict(input_data_test)

# Evaluate the model's performance
r2_score = r2_score(output_data_test, y_pred)
print('R2 score Linear Regression:', 100*r2_score)
```

### **8.2 Optimization technique using Randomised Search Cv for SVR**
